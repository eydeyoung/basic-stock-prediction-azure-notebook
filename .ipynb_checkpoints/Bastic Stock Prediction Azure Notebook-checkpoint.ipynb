## Problem Statement

Investors are interested in the capability to predict stock prices and returns given a priori information or information on other stocks as this enables them to enact strategies (buy low/sell high) that can allow them to make money trading a stock or index.

This notebook concerns itself with the prediction of the one day returns of the Istanbul Stock Exchange index (ISE) using the following stock indices from January 2009 to February 2011:

- ISE
- S&P
- DAX
- FTSE
- NIKKEI
- BOVESPA
- EU
- EM

In particular it is of interest to see whether linear regression models are more suitable and comparable to other nonlinear regression models such as a neural network, or to a SARIMAX type model.



## Loading the data

```python
# data source
# https://archive.ics.uci.edu/ml/datasets/ISTANBUL+STOCK+EXCHANGE

df = pd.read_excel("data_akbilgic.xlsx", skiprows = [0])
df.head()
```

The notebook contains several ETF indices as the columns. Each row contains the one day simple(percent) return of the adjusted close of the stocks from January 2009 to February 2011. We shall use this data to determine the suitability of our models for that time period

```python
#Let's rename the columns to reflect that ISE is TL based and ISE.1 is USD based
df.rename(columns = {"ISE": "ISE_TL", "ISE.1": "ISE_USD"}, inplace = True)
```

```python
#Change date to date time and set as index

df["date"] = pd.to_datetime(df["date"] )
df.set_index(df["date"], inplace = True)

try:
  df.drop("date", axis = 1, inplace = True)
except:
  pass
```

### Plotting the Decomposition of the ISE

```python
res = sm.tsa.seasonal_decompose(df["ISE_USD"], period = 52)
fig = res.plot()
fig.set_figheight(15)
fig.set_figwidth(15)
plt.show()
```

We plot the seasonal decomposition of the ISE timeseries and note that the residuals appear randomn and there appears to be some seasonality in the data (period appears to be slightly under 3 months).

### Plotting the AutoCorrelation

```python
#plot autocorrelation to determine if past returns are correlated
smt.graphics.plot_acf(df["ISE_USD"], lags = 30)
plt.title("AutoCorrelation of ISE ")
plt.xlabel("Lags")
plt.show()
```

We see there is very little autocorrelation of the time series. The past returns of the ISE are typically not very informative. The thirteenth, twenty-seventh and the thirtyeth are the only somewhat significant lags. As even the immediate returns have very low autocorrelation there is no significant trend.

This supports the hypothesis that the series is very stationary and mean reverting. Hence the price will tend back towards an average and there is a chance to make money on a buy low sell high strategy.

We can see that the cyclicity in the data is present over a thirty day period but it is very weak in comparison to the autocorrelation of the lag in general. We move on to the partial autocorrelation function to see if it supports these results.

```python
smt.graphics.plot_pacf(df["ISE_USD"], lags = 31)
plt.title("Partial AutoCorrelation of ISE ")
plt.xlabel("Lags")
```

The partial autocorrelation function is very similar to the autocorrelation function hence there is little more information to be gleaned from it.

## Preparation

### Checking if the Data Needs Any Cleaning

```python
df.describe()
```

```python
print(df.isna().sum())
print(df.dtypes)
```

There are no nan entries in the dataset. Each column only consists of floating point entries and has a reasonable minimum and maximum based on the standard deviation.

The data can therefore be expected to be valid and free of signifacant outliers, so there is little cleaning to be done in the preparation step, save for one thing...



In the general theory of price, returns are lognormal which means the log of the returns follows a normal distribution. For linear regression and the like, the log returns will be more informative than the percent returns as they support the normality assumption.


```python
#Convert Percent Returns to Log Returns:

for col in df.columns:
  df[col] = np.log(1 + df[col] )
```

### Feature Engineering 

The prior data exploration showed that there was very little momentum and trending. 

There was very little autocorrelation but the thirtheenth, twenty-seventh and thirtieth lagged returns showed some potential significance. In addition the season decomposition showed there may be some slight seasonality.

Based on this, there may be some benefit to adding a small window feature looking immediately at the past two days average log returns as well as the three significant lags.

```python
#Generate a rolling window mean of the only the past two work days

df["rolling_2_mean"]= df["ISE_USD"].rolling(2).mean()
```

```python
#Generating Lagged returns:

for i in [13,27,30]:
  df["ISE_USD_lag_" + str(i)] = df["ISE_USD"].shift(i)

```

To generate our target we shift the ISE stock exchange up by one so we are predicting the next days return based on the previous days return, the current stock indices values

```python
df["ISE_USD_target"] = df["ISE_USD"].shift(-1)

```

```python

#Drop the rows with the nan entries from missing data in the 13 day returns
#and from shifting
df.dropna(subset = ["rolling_2_mean","ISE_USD_target", "ISE_USD_lag_30"], axis = 0, inplace = True) 
```

```python
df.head()
```

### Splitting the data

As the data is a timeseries the order of points matters, so we can only split our training and testing data without shuffling randomly as is done in train_test_split.

As our dataset is fairly small we split our dataset into 90% training and 10% testing

```python
test_percent = 0.1

train_end_index = int(df.shape[0]*(1 - test_percent))

train = df.iloc[:train_end_index, :]
test = df.iloc[train_end_index:, :]
```

#### Scaling the Data

```python
#Scale the data to have unit variance and 0 mean
scaler = StandardScaler()
scaler.fit(train)

train = pd.DataFrame(scaler.transform(train), index = train.index, columns = train.columns)
test = pd.DataFrame(scaler.transform(test), index = test.index, columns = test.columns)
```

```python
#Drop information leakage variations of the target
#Drop the excessive highy correlated features DAX and the FTSE
X_train = train.drop(["ISE_TL","ISE_USD","DAX", "FTSE", "ISE_USD_target"], axis = 1)
y_train = train["ISE_USD_target"]

X_test = test.drop(["ISE_TL","ISE_USD","DAX", "FTSE", "ISE_USD_target"], axis = 1)
y_test = test["ISE_USD_target"]
```

## Basic Models

We consider a linear regressor model and a decision tree model.

The linear regressor model is suitable if there is a sufficient linear trend that explains the variation in the data.

As our data has a weak trend, adding in the lags may not be enough to make a linear regressor suitable. Other regressors such as a KNN regressor, decision tree regressor or a gradient boosting regressor may be more suitable to forecast the stock returns.

### Lasso

```python
lr = Lasso(random_state = 1, max_iter = 2000)
lr.fit(X_train, y_train)

predictions = lr.predict(X_test)
rmse = math.sqrt(mean_squared_error(y_test, predictions))
print("Basic Lasso RMSE is : %.5f" % rmse)
```

### Neural Net

```python
nn = MLPRegressor(random_state = 1)
nn.fit(X_train, y_train)

predictions = nn.predict(X_test)
rmse = math.sqrt(mean_squared_error(y_test, predictions))
print("Basic Neural Net RMSE is : %.5f" % rmse)
```

### Decision Tree Regressor

```python
dt = DecisionTreeRegressor(random_state = 1)
dt.fit(X_train, y_train)

predictions = dt.predict(X_test)
rmse = math.sqrt(mean_squared_error(y_test, predictions))
print("Basic Decision Tree RMSE is : %.5f" % rmse)
```

The basic neural network has the best performance. This may signify that a nonlinear method may be the best predictor, as opposed to a linear regressor like Lasso. The decision tree regressor has a relatively high error in comparison to the other methods which may be due to a loss of information in the grouping and splitting inherent to the method.

## Hyperparameter Tuning

### Lasso

```python
#Need to use a timeseries splitter so our cross validations are valid for the
#time series

timesplit = TimeSeriesSplit()

lr = Lasso(random_state = 1, max_iter = 2000 )
params = {"alpha":[0.0001, 0.01,0.1, 1, 10, 100, 10000]}

gridsearch = GridSearchCV(estimator = lr, param_grid = params, scoring = "neg_mean_squared_error", cv = timesplit)

gridsearch.fit(X_train, y_train)

bestmodel = gridsearch.best_estimator_
```

```python
print(gridsearch.best_params_)
print(bestmodel.coef_)
```

```python
lasso_predictions = bestmodel.predict(X_test)
print("Tuned Lasso RMSE is : %.5f" % math.sqrt(mean_squared_error(y_test, lasso_predictions)))
```

The best Lasso model is one with low regulation strength, such so that it cannot drive any of the remaining features down to 0 as is typically the result of Lasso's sparse coefficient nature. The BOVESPA and S&P appear as the most relevant features contributing to an increase and the EM index the most relevant that has a negative effect.

### MLP Based

```python
# Takes about 6 minutes
nn = MLPRegressor(random_state = 1, max_iter = 500)

nn_params = {"hidden_layer_sizes":[(100,), (200,), (100,100,), (200,200,) \
                                   , (100,100,100), (200,200,200)] \
             , "activation":["identity", "logistic", "tanh", "relu"] \
             , "alpha":[0.0001, 0.001, 0.01]
             , "learning_rate_init": [0.01, 0.1]}

gridsearch = GridSearchCV(estimator = nn, param_grid = nn_params, scoring = "neg_mean_squared_error", cv = timesplit)

gridsearch.fit(X_train, y_train)

bestmodel = gridsearch.best_estimator_
```

```python
print(gridsearch.best_params_)
```

```python
nn_predictions = bestmodel.predict(X_test)
print("Tuned Neural Network RMSE is : %.5f" % math.sqrt(mean_squared_error(y_test, nn_predictions)))
```

The best homogenous activation neural network is one with an identity activation that makes it effectively a linear function similar to the Lasso so we can say that the linear regression may be the most suitable for making predictions in this dataset for the time period of 2009 - 2011.

### Decision Tree Regressor

```python
dt = DecisionTreeRegressor(random_state = 1)

dt_params = {"criterion":["squared_error", "friedman_mse", "absolute_error", "poisson"] \
             , "max_depth":[5, 10, None] \
             , "min_samples_split":[2,5,10]}

gridsearch = GridSearchCV(estimator = dt, param_grid = dt_params, scoring = "neg_mean_squared_error", cv = timesplit)

gridsearch.fit(X_train, y_train)

bestmodel = gridsearch.best_estimator_
```

```python
print(gridsearch.best_params_)
```

```python
dt_predictions = bestmodel.predict(X_test)
print("Tuned Decision Tree RMSE is : %.5f" % math.sqrt(mean_squared_error(y_test, dt_predictions)))
```

The decision tree error has dramatically decreased with hyperparameter tuning and now approached the errors of the Lasso linear regression and neural network. Overall the best methods seem to be the linear approximation methods when the lags are included.

```python
results = np.column_stack( (y_test.values, lasso_predictions, nn_predictions, dt_predictions))

results_df = pd.DataFrame(results, index = y_test.index \
                          , columns = ["target", "lasso_pred" \
                                       , "nn_pred", "dt_pred"])
results_df
```

## Conclusions

A look at the results shows the Lasso and Neural Network models effectively end selecting a small subset of values with which they predict the actual value hence the accuracy of their regression is very limited. These results indicate that either more data is required or that the models would be better suited to predicting the classifications of the stock returns (positive or negative) rather than the actual return values. This would still prove useful for a trading strategy based on the mean reversion of the stocks.
